```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Salary_Data.csv')
print(data.head(10))
```

```python
data.describe()
```

```python
plt.title('Salary Distribution Plot')
sns.distplot(data['Salary'])
plt.show()
```

```python
plt.scatter(data['YearsExperience'], data['Salary'])
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()



```

```python
# Relationship between Salary and Experience
plt.scatter(data['YearsExperience'], data['Salary'], color = 'lightcoral')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(False)
plt.show()
```

```python
x = data.iloc[:, :-1].values  # Features (Years of Experience)
y = data.iloc[:, 1].values    # Target variable (Salary)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
regressor = LinearRegression()      
regressor.fit(x_train, y_train)  # Fit the model on training data
y_pred_test = regressor.predict(x_test)  # Predict on test data  
y_pred_train = regressor.predict(x_train)  # Predict on training data
```

```python
plt.scatter(x_train, y_train, color = 'lightcoral')
plt.plot(x_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(x_test, y_test, color = 'lightcoral')
plt.plot(x_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred_test)
print("Mean Absolute Error (MAE):", mae)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error (MSE):", mse)
```

```python
import numpy as np

# Pick a random index from the test set
random_idx = np.random.randint(0, len(y_test))
actual = y_test[random_idx]
predicted = y_pred_test[random_idx]
error = actual - predicted

print(f"Random Error for test sample {random_idx}: Actual = {actual}, Predicted = {predicted}, Error = {error}")
```
