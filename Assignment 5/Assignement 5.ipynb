```python

from sklearn import datasets
cancer = datasets.load_breast_cancer()
```

```python
print(cancer.data)
```

```python
print("Features: ", cancer.feature_names)
```

```python
print("Labels: ", cancer.target_names)
```

```python
cancer.data.shape
```

```python
print(cancer.data[0:5])
```

```python
print(cancer.target)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, test_size=0.3, random_state=109
)
```

```python
from sklearn import svm
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
# Cell 11
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=cancer.target_names,
            yticklabels=cancer.target_names)
plt.title('Confusion Matrix for Breast Cancer Prediction', pad=20)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Calculate and display accuracy from the confusion matrix
accuracy = (cm[0, 0] + cm[1, 1]) / np.sum(cm)
plt.figtext(0.5, -0.05, 
            f'Total Samples: {len(y_test)}   |   Accuracy: {accuracy:.2%}', 
            ha="center", fontsize=10)

# Optionally, extract TN, FP, FN, TP values:
tn, fp, fn, tp = cm.ravel()

plt.tight_layout()
plt.show()
```
